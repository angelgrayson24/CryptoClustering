```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics # this is to look at the data 
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler().fit_transform(df_market_data)
scaler


```

```python

# Select relevant columns for standardization
features_to_scale = ['price_change_percentage_24h', 'price_change_percentage_7d', 'price_change_percentage_14d', 'price_change_percentage_30d', 'price_change_percentage_60d', 'price_change_percentage_200d', 'price_change_percentage_1y']

# Extract the selected columns
X = df_market_data[features_to_scale]

# Use StandardScaler to normalize the columns
scaler = StandardScaler()
scaled_features = scaler.fit_transform(X)

# Create a new DataFrame with the scaled features
scaled_df = pd.DataFrame(scaled_features, columns=features_to_scale, index=df_market_data.index)

# Copy the crypto names from the original data
og_names = list(df_market_data.index)
scaled_df['coin_id'] = og_names

# Set the 'coin_id' column as index
scaled_df.set_index('coin_id', inplace=True)

# Display sample data
scaled_df
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
k_values
```

```python
# Create an empty list to store the inertia values


# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list


# Select relevant features for clustering
features_for_clustering = ['price_change_percentage_24h', 'price_change_percentage_7d', 'price_change_percentage_14d', 'price_change_percentage_30d', 'price_change_percentage_60d', 'price_change_percentage_200d', 'price_change_percentage_1y']
df_market_data_scaled = StandardScaler().fit_transform(df_market_data[features_for_clustering])

# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
for k in range(1, 12):
    # Create a KMeans model using the loop counter for the n_clusters
    kmodel = KMeans(n_clusters=k, random_state=42)
    
    # Fit the model to the data using `df_market_data_scaled`
    kmodel.fit(df_market_data_scaled)
    
    # Append the model.inertia_ to the inertia list
    inertia_values.append(kmodel.inertia_)

# Display the inertia values
print(inertia_values)
```

```python
# Create a dictionary with the data to plot the Elbow curve
inertia_dict = {}

# Create a for loop to compute the inertia with each possible value of k
for k in range(1, 12):
    # Create a KMeans model using the loop counter for the n_clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    
    # Fit the model to the data using `df_market_data_scaled`
    kmeans.fit(df_market_data_scaled)
    
    # Add the inertia value to the dictionary
    inertia_dict[k] = kmeans.inertia_

# Display the inertia dictionary
print(inertia_dict)

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(list(inertia_dict.items()), columns=['k', 'Inertia'])

# Display the DataFrame
elbow_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_df.hvplot.line(
    x="k",
    y="Inertia",
    title="Elbow Curve",
    xlabel='Number of Clusters (k)',
    ylabel='Inertia',
    xticks=list(range(1, 12))
)

```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** k = 4
